Importing Datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading track dataset
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

Spotify Tracks Datasets Analysis

```python
# null value
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

Descriptive statistics of Spotify tracks

```python
df_tracks.describe().transpose()
```

Top 5 most popular songs on Spotify

```python
most_poplular_songs = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_poplular_songs[:5]
```

Top 5 least popular songs on Spotify

```python
least_popular_songs = df_tracks.sort_values('popularity', ascending = True)
least_popular_songs[:5]
```

Changing index to release date

```python
df_tracks.set_index('release_date', inplace = True)
df_tracks.index=pd.to_datetime(df_tracks.index, format='ISO8601')
df_tracks.head()
```

Check artist at 18th row of Spotify Tracks Dataset

```python
df_tracks[['artists']].iloc[18]
```

Converting duration from millisecond to second on Spotify Tracks Dataset

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x : round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
df_tracks.duration.head()
```

Considering 0.4% of the total dataset to create Regression plots

```python
sample = df_tracks.sample(int(0.004 * len(df_tracks)))
len(sample)
```

Regression plot - Correlation between Loudness and Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample, y='loudness', x='energy', color='c').set(title='Loudness vs Energy Correlation')
```

Regression plot - Correlation between Popularity and Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample, y='popularity', x='acousticness', color='b').set(title='Popularity vs Acousticness Correlation')
```

Creating new column in tracks dataset (Year, Release Date)

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
df_tracks.head()
```

Distribution plot - Visualize total number of songs on Spotify since 1992

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number of songs per year')
```

Change in Duration of songs wrt Years

```python
total_dr = df_tracks.duration
sns.set_style(style = 'whitegrid')
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax, err_kws={'linewidth': False}).set(title = 'Year vs Duration')
plt.xticks(rotation = 60)
```

Spotify Features Dataset Analysis

```python
df_genre.head()
```

Duration of songs in different Genres

```python
plt.title('Duration of songs in different Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data = df_genre)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

Top 5 Genres by Popularity

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10, 5))
famous = df_genre.sort_values('popularity', ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = ' Top 5 Genres by Popularity')
```
